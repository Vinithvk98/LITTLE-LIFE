```python
X = dataset.drop('fetal_health',axis='columns')
#print(X.shape)
#X
Y = dataset.fetal_health
```

```python
#feature selection using stats Model
import statsmodels.api as sm
X = sm.add_constant(X)
```

```python
lr = sm.OLS(Y,X).fit()
print(lr.summary2())
```

```python
p_values = lr.pvalues
```

```python
vars = p_values[p_values<=0.05].index.tolist()
```

```python
print(vars)
```

```python
X = X[vars[1:]]
```

```python
X.head()
```

```python
Y.head()
```

#**Solving Class Imbalance**

**Oversampling using SMOTE**

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_sm, y_sm = smote.fit_resample(X, Y)
```

```python
X_sm.head()
```

```python
y_sm.head()
```

```python
y_sm.value_counts()
```

```python
fig = plt.figure(figsize=(10,6))
ax = (y_sm.value_counts()*100.0 /len(y_sm)).plot.pie(autopct='%.1f%%',fontsize=12)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_sm, y_sm, test_size = 0.3, stratify= y_sm, random_state = 42)
```

# Data Normalization

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Finding best max_depth Value

accuracy = []
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

for i in range(1, 15):
    model = DecisionTreeClassifier(max_depth = i, random_state = 42)
    model.fit(X_train, Y_train)
    pred = model.predict(X_test)
    score = accuracy_score(Y_test, pred)
    accuracy.append(score)

plt.figure(figsize=(12, 6))
plt.plot(range(1, 15), accuracy, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Finding best Max_Depth')
plt.xlabel('pred')
plt.ylabel('score')
plt.show()
```


#**DECISION TREE**
